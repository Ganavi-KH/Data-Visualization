# Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.arange(1,8)
x
y=x*2
y
plt.plot(x,y, color='green', marker='h', markerfacecolor='blue', markeredgecolor='red',
         markersize='10', markeredgewidth='1', linestyle='-.', linewidth=2, label='num')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph')
z=x*3
plt.plot(x,z, color='blue', marker='*', ms='10', mfc='deeppink', mec='green', mew='1', ls=':', lw='2', label='num*3')
plt.legend(loc='upper left', facecolor='blue', edgecolor='black', framealpha=0.5)
plt.grid(axis='both', ls=':', lw='2', color='olive')
plt.show()
```

```python
plt.subplot(3,2,1)
plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph')

plt.subplot(3,2,2)
plt.plot(x,y*2)

plt.subplot(3,2,3)
plt.plot(x,y*3)

plt.subplot(3,2,4)
plt.plot(x,y*4)

plt.subplot(3,2,5)
plt.plot(x,y*5)

plt.subplot(3,2,6)
plt.plot(x,y*6)

plt.suptitle('Main plot', fontsize=18)
plt.tight_layout()
```

```python
fig, axes=plt.subplots(3,2, facecolor='plum', sharex=True, sharey=True)
x=np.linspace(1, 10, 2)
y=x**2
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('x-axis')
axes[0][0].set_ylabel('y-axis')
axes[0][0].set_facecolor('pink')
axes[0][0].set_title('lineplot', fontsize=14)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(x,y)
axes[2][0].plot(x,y)
axes[2][1].plot(x,y)

fig.suptitle('Main plot', fontsize=18)
fig.tight_layout(h_pad=2, w_pad=2)
plt.show()
```

# Pie Chart

# Stacked Bar Chart

```python
df1.columns= df1.columns.str.strip().str.lower()

unit=['Finance', 'IT', 'Management', 'Marketing', 'Operations', 'Web']
u_count=df1['unit'].str.strip().value_counts()
u_total=[u_count.get(u, 0) for u in unit]

Designation=['Analyst', 'Associate', 'Director', 'Manager', 'Senior Analyst', 'Senior Manager']
d_counts=df1['designation'].str.strip().value_counts()
Des_total= [d_counts.get(d, 0) for d in Designation]

year= [2010, 2011, 2012, 2013, 2014, 2015]
yr_count=df1['year'].value_counts()
yr_total=[yr_count.get(yr, 0) for yr in year]

b_yr= [i+j for i,j in zip(u_total, Des_total)]

plt.figure(figsize=(8,5))
plt.bar(unit, u_total, color='yellow', label='Unit', width=0.5)
plt.bar(unit, Des_total, color='orchid', bottom=u_total, label='Designation', width=0.5)
plt.bar(unit, yr_total, color='limegreen', bottom=b_yr, label='Year', width=0.5)
plt.xlabel('Unit')
plt.ylabel("Employee's working in Units")
plt.title('Count of Employees by Units')
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylim(0,3200)
plt.legend(loc='upper left', facecolor='blue', edgecolor='black', framealpha=0.3)
plt.show()
```

# Grouped Bar Chart

```python
unit=['Finance', 'IT', 'Management', 'Marketing', 'Operations', 'Web']
u_count=df1['unit'].str.strip().value_counts()
u_total=[u_count.get(u, 0) for u in unit]

Designation=['Analyst', 'Associate', 'Director', 'Manager', 'Senior Analyst', 'Senior Manager']
d_counts=df1['designation'].str.strip().value_counts()
Des_total= [d_counts.get(d, 0) for d in Designation]

year= [2010, 2011, 2012, 2013, 2014, 2015]
yr_count=df1['year'].value_counts()
yr_total=[yr_count.get(yr, 0) for yr in year]

w=0.2
st= [i for i in range(len(unit))]
st1= [i+w for i in st]
st2= [i+w for i in st1]
st=np.array(st)

plt.figure(figsize=(8,5))
plt.xticks(st+w+0.1, unit, rotation=45)
f={'size':16, 'weight':'bold', 'color':'teal', 'style':'italic'}

plt.bar(st, u_total, width=w, color='blue', label='Unit')
plt.bar(st1, Des_total, width=w, color='limegreen', label='Designation')
plt.bar(st2, yr_total, width=w, color='deeppink', label='Year')

plt.xlabel('Unit', fontdict=f)
plt.ylabel("Employee's working in Units", fontdict=f)
plt.title('Count of Employees by Units', fontdict=f)
plt.legend(facecolor='blue', edgecolor='black', framealpha=0.3)
plt.show()
```
