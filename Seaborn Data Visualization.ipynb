# Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ITWorkforce.csv')
df
```

# Count Plot

```python
sns.countplot(x='Gender', data=df, hue='Race', palette='twilight')
```

```python
sns.countplot(x='Gender', data=df, hue='Race', palette='cividis', order=['Female', 'Male'],
       hue_order=['White', 'Hispanic', 'Asian', 'Korean', 'Mixed', 'Black', 'Chinese', 'Australian', 'Welsh', 'African American'])
```

```python
sns.countplot(y='Gender', data=df, hue='Race', palette='Spectral', order=['Female', 'Male'],
       hue_order=['White', 'Hispanic', 'Asian', 'Korean', 'Mixed', 'Black', 'Chinese', 'Australian', 'Welsh', 'African American'])
```

# Bar Plot

```python
sns.barplot(x='Education Level', y='Age', data=df, hue='Gender', palette='YlGnBu', estimator=np.sum, 
            hue_order=['Female', 'Male'], order=[3,2,1,0])
```

# Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Senior', y='Education Level', data=df, hue='Country', palette='YlOrBr', showmeans=True,
            meanprops={'marker':'o', 'markerfacecolor':'red', 'markeredgecolor':'white', 'markersize':10})

Senior=df['Senior'].dropna()
Q1=Senior.quantile(0.25)
Q2=Senior.quantile(0.50)
Q3=Senior.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

# Violin Plot

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Country', y='Years of Experience', data=df, hue='Gender', palette='inferno', split=True)
```

# Distribution plot

# Hist  Plot

```python
sns.histplot(x='Age', data=df, hue='Education Level', palette='bright', kde=True, multiple='dodge')
```

```python
sns.displot(x='Age', data=df, hue='Education Level', palette='bright', kde=True, multiple='dodge', 
            col='Education Level', row='Senior', height=3, aspect=1)
```

# Join Plot

```python
sns.jointplot(x='Years of Experience', y='Salary', data=df, kind='reg', palette='Purples')
```

```python
sns.jointplot(x='Years of Experience', y='Salary', data=df, kind='reg', palette='Purples',
              joint_kws={'marker':'*', 'color':'purple', 'scatter_kws': {'s':50,}}, 
              marginal_kws={'color':'red', 'bins':10}, marginal_ticks=True)
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, vars=['Age', 'Gender', 'Education Level'])
```

```python
sns.pairplot(df, vars=['Age', 'Education Level'], hue='Senior', palette='plasma', height=3, aspect=1)
plt.show()
```

# Heat Map

```python
correlation_matrix= df.corr(numeric_only=True)
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='magma', linewidths=0.1, linecolor='black')
```
